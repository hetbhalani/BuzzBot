```python
from langgraph.graph import START, END, StateGraph
from pydantic import BaseModel
from typing import TypedDict, Optional
import datetime
```

```python
class DailyPipelineState(TypedDict):
    today: str = datetime.datetime.now().strftime("%A")
    
    raw_news: list[dict]
    enriched_news_w_rank: list[dict]
    ranked_news: list[dict]
    
    s3_key: Optional[str]
    number_of_news: Optional[int]
    
    errors: list[str]
    
```

```python
class WeeklyPipelineState(TypedDict):
    today: str = datetime.datetime.now().strftime("%A")
    
    all_news: list[dict]
    top_news: list[dict]
    
    draft_post: Optional[str]
    post_word_count: Optional[int]
    
    approve_status: Optional[str]
    user_prompt: Optional[str]
    edited_response: Optional[str]
```
